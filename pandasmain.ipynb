```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Calculate the number of unique schools in the DataFrame
totschoolcount = len(school_data_complete["school_name"].unique())
totschoolcount
```

```python
pass_math = sum(student_data["math_score"]>=70)/totstudents*100
pass_math
```

```python
pass_read = sum(student_data["reading_score"]>=70)/totstudents*100
pass_read
```

```python
passoverall_df = student_data[(student_data["reading_score"]>=70) & (student_data["math_score"]>=70)]
passoverall_df
```

```python
passoverall = (passoverall_df.count()[0])/totstudents*100
passoverall
```

```python
pass_read_df = (student_data.loc[student_data["reading_score"]>=70, :]) 
pass_read_df
```
