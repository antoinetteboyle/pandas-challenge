```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Print the first five rows of data to the screen
student_data.head()
```

```python
school_data.head()
```

```python
school_data_complete.head()
```

# DISTRICT SUMMARY:

```python
# Calculate the number of unique schools in the DataFrame
totschoolcount = len(school_data_complete["school_name"].unique())
totschoolcount
```

```python
# Calculate the number of unique students in the DataFrame
totstudents = len(school_data_complete["Student ID"].unique())
totstudents
```

```python
# Calculate the Total Budget
totalbudgetcount = school_data["budget"].sum()
totalbudgetcount
```

```python
# Calculate the Average Math
average_math = student_data["math_score"].mean()
average_math
```

```python
# Calculate the Average Read
average_read = student_data["reading_score"].mean()
average_read
```

```python
# Calculate the % passing Math
pass_math = sum(student_data["math_score"]>=70)/totstudents*100
pass_math
```

```python
# Calculate the % passing Reading
pass_read = sum(student_data["reading_score"]>=70)/totstudents*100
pass_read
```

```python
# Calculate the % passing overall
passoverall_df = student_data[(student_data["reading_score"]>=70) & (student_data["math_score"]>=70)]
passoverall_df
```

```python
passoverall = (passoverall_df.count()[0])/totstudents*100
passoverall
```

```python
# Place all of the data found into a summary "DISTRICT" DataFrame
summary_df = pd.DataFrame({"Total Schools": [totschoolcount],
                              "Total Students": totstudents,
                              "Total Budget": totalbudgetcount,
                              "Average Math Score": average_math,
                              "Average Reading Score": average_read,
                              "% Pass Math": pass_math,
                              "% Pass Reading": pass_read,
                              "% Overall Pass": passoverall,})
summary_df
```

# SCHOOL SUMMARY:

```python
school_df = school_data_complete.groupby(['school_name',"type"])['math_score','reading_score'].agg('mean')
#school_df = pd.DataFrame(school_df)
school_df.head()
```

```python
school_df = school_df.reset_index()
school_df = school_df.set_index("school_name")
school_df
```

School Summary: Calculate Budget per student, save to variables and add to datafram

```python
students = school_data_complete["school_name"].value_counts()
students
```

```python
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_budget
```

```python
per_student_budget = per_school_budget / students
per_student_budget
```

```python
school_df["School Budget"] = per_school_budget
school_df["Students"] = students
school_df["Student Budget"] = per_student_budget

school_df
```

School Summary: Calculate % pass in Math and Reading and Overall and add to dataframe

```python
school_pass_math = school_data_complete.loc[(school_data_complete["math_score"]>=70),['school_name','math_score']]
math = school_pass_math["school_name"].value_counts()
math
```

```python
mathperc = math/students*100
mathperc
```

```python
school_pass_read = school_data_complete.loc[(school_data_complete["reading_score"]>=70),['school_name','reading_score']]
reading = school_pass_read["school_name"].value_counts()
reading
```

```python
readperc = reading/students*100
readperc
```

```python
school_pass_overall = school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70) ,['school_name','math_score']]
overall = school_pass_overall["school_name"].value_counts()
overall
```

```python
overallperc = overall/students*100
overallperc
```

```python
school_df["% Pass Math"] = mathperc
school_df["% Pass Reading"] = readperc
school_df["% Pass Overall"] = overallperc
school_df
```

```python
school_df = school_df.rename(columns = {'math_score':'Average Math Score', 'reading_score':'Average Reading Score'})
school_df
```

# BUDGET PER STUDENT BINS

```python
school_df.head()
```

```python
studentbudgetbin = school_df.reset_index()
del studentbudgetbin['School Budget']
del studentbudgetbin['Students']
del studentbudgetbin['type']
studentbudgetbin
```

```python
# Create the bins in which Data will be held
bins = [0, 585, 630, 645, 680]
```

```python
# Create the names for the bins
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
studentbudgetbin["Student result Summary"] = pd.cut(studentbudgetbin["Student Budget"], bins, labels=group_names, include_lowest=True)
studentbudgetbin
```

```python
studentbudgetbin.groupby(['Student result Summary'])[['Average Math Score','Average Reading Score','% Pass Math','% Pass Reading','% Pass Overall']].agg(['mean'])
```
